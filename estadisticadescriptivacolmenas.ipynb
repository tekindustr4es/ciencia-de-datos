Por último, se presenta un diagrama de cajas y bigotes, para la temperatura y la humedad

```python
import matplotlib.pyplot as plt
import numpy as np

# Generamos datos aleatorios

datos1 = Temperatura
datos2 = Humedad


# Creamos una lista con los datos
datos = [datos1, datos2]

# Creamos la figura y los ejes
fig, ax = plt.subplots()

# Creamos el diagrama de cajas y bigotes
ax.boxplot(datos)

# Añadimos etiquetas a los ejes
ax.set_xticklabels(['Temperatura', 'Humedad'])

# Añadimos un título al gráfico
ax.set_title('Diagrama de Cajas y Bigotes de la Temperatura y Humedad de las Colmenas de Abejas')

# Mostramos el gráfico
plt.show()
```
